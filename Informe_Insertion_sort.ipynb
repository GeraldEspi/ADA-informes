# 2.- Insertion sort

El siguiente código muestra la implementación de un **Insertion Sort** en python, para una mejor visualización ignorar lo encerrado entre líneas verdes.

# 2.1. Código

```python
  from termcolor import colored
  import copy

  def insertionSort(array, verbose = False):
    cont = 0
    cont2 = 1
    for sp in range(1, len(array)):
      #--------------parte del verbose--------------- 

        if(verbose == True):
            print(colored('Iteración número ','blue'),colored(cont2,'green'),'\n')
      #-----------------------------------------------
        key = array[sp]
        j = sp - 1
        #--------------parte del verbose--------------- 
        cont3 = 0
        if(key > array[j]):
          if(verbose == True and cont3 == 0):
                print(colored('#Se compara#\n','green'))
                for i in range(0,len(array)+1):
                    if (i == 0):
                      print(' [ ',end = '')
                    if(i == len(array)):
                        if(array[i-1] == key or array[i-1] == array[j]):
                          print(colored(array[i-1],'blue'),']',end = '')
                        else:
                          print(array[i-1],']',end = '')
                      
                    if(i < len(array)-1):
                        if(array[i] == key or array[i] == array[j]):
                          print(colored(array[i],'blue'),end = ', ')
                        else:
                          print(array[i],end = ', ')
                print('\n')
          cont += 1 
         
        #-----------------------------------------------

        while j >= 0 and key < array[j]:
          #--------------parte del verbose--------------- 
            if(verbose == True and cont3 == 0):
                print(colored('#Se compara#\n','green'))
                for i in range(0,len(array)+1):
                    if (i == 0):
                      print(' [ ',end = '')
                    if(i == len(array)):
                        if(array[i-1] == key or array[i-1] == array[j]):
                          print(colored(array[i-1],'blue'),']',end = '')
                        else:
                          print(array[i-1],']',end = '')
                      
                    if(i < len(array)-1):
                        if(array[i] == key or array[i] == array[j]):
                          print(colored(array[i],'blue'),end = ', ')
                        else:
                          print(array[i],end = ', ')
            
                print('\n')
          #-----------------------------------------------
            array[j + 1] = array[j]
            cont += 1 
            cont3 += 1         
            j = j - 1
 

        array[j + 1] = key


        #--------------parte del verbose--------------- 
        if(verbose == True ):
         print(colored('#El número','green'),key,colored('se mueve a la posición correspondiente: ','green'),array,'\n') 
        cont2+=1
        #-----------------------------------------------
      

                
    
    return cont,array


data = [9, 5, 1, 4, 3]
print(colored('Arreglo de entrada:', 'green'),data,'\n')
cont, a= insertionSort(data)
print(colored('Arreglo ordenado en orden ascendente:','green'),data)
print(colored('\nComparaciones:', 'green'), cont,'\n')


```

## 2.2. Descripción del algoritmo

El algoritmo recibe una lista (arreglo) de datos sin un orden específico. Esta lista contiene una cantidad de números **N**. Luego de ingresada a la función, el algoritmo ordena la lista de menor a mayor.

**Explicación paso a paso:**

1.- En cada iteración de la variable **sp** se compara la `key` asociada hasta que se encuentre un número con valor **menor**. Si se quisiera ordenar de manera descendente se debería cambiar en el ciclo while `key<array[j]` a `key>array[j]` y de esta manera, buscaría un número con valor **mayor**.

2.- Una vez encontrado el número se posiciona la variable `key` inmediatamente después del elemento encontrado.

3.- luego de completadas las iteraciones correspondientes, se retorna un contador y el arreglo ya ordenado.



# 2.3. Ejemplo

Se considera un arreglo con 4 variables:

**À** = **[7 , 4 , 8 , 3]**

En la primera iteración se establece la key, en este caso, como la key corresponde a `key = array[sp]` correponderá a array[1], siendo el número 4. Se mueve a la primera posición y los demás números se mueven a la izquierda en 1 unidad.

**À** = **[4 , 7 , 8 , 3]**

Luego la key corresponderá al siguiente número del arreglo de la posición, si antes era `key = array[sp]` ahora key corresponde a 
`key = array[sp+1]` en este caso 8. Luego se compara con los anteriores, en este caso no hay que hacer cambios, está ordenado.

**À** = **[4 , 7 , 8 , 3]**

El siguiente número más pequeño es 3 por lo que este se compara con los demás hasta llegar a la primera posición del arreglo por ser el más pequeño.

**À** = **[3 , 4 , 7 , 8]**

Finalmente se comprueba que el orden sea correcto a través de una ultima pasada y se finaliza el proceso.

Al ser de complejidad ***$O(n^2)$*** el número de comparaciones se calcula como $\sum\limits_{i=0}^{n-1} i=\frac{n(n-1)}{2}$

# 4. Correctitud
### **Teorema (Correctitud).**

*El algortimo insertion sort genera un arreglo con los elementos de entrada ordenados de menor a mayor o mayor a menor según se le indique.*

Entrada: **$A[ 1, ...  j'-1 ]$**

Los elementos del arreglo se encuentran ordenados de menor a mayor.

## Prueba del Teorema

Se pudo comprobar con anteriorioridad que **no** en todas las iteraciones se hacen cambios en el arreglo en cuestión. A pesar de ello mientras la key avanza y se dan nuevas y diferentes comparaciones es posible visualizar como los datos se ordenan mientras más iteraciones ocurren. Para comprobar su correctitud se hará uso la **propiedad invariante de bucle** para `Insertion Sort`.

> Al comienzo de cada iteración del bucle for se establece `key` y se encuentra un número menor al correspondido por `key`, si no se logra encontrar uno, key terminará aumentando, eventualmente un dato del arreglo se posicionará correctamente.

**Inicialización**

Se tiene que $j' = 1$ por lo tanto, **$A[ 1, ...  j'-1 ]$** = **$A[1]$** un arreglo de solo una variable siempre está ordenado.

**Mantención**

Para una $j$ particular

**$A[ 1, ...  j'-1 ]$** está ordenado mientras que el blucle posicione a **$A[j']$** en la posición correcta. 
Significando que **$A[ 1, ...  j'-1 ]$** está ordenado.

En la siguiente iteración **$j$** corresponde a **$j'+1$** lo que siginifica que
**$A[ 1, ...  j'-1 ]$** está ordenado, lo que a su vez indica que la invariable de bucle sigue siendo verdadera.

**Correctitud**

El ciclo 'for' terminará cuando $j'>n$, (es decir,  $j' = n+1$)
el subarreglo basado en la invariable de bucle **$A[ 1, ...  n ]$**, por definición está correctamente ordenado. Por lo tanto, el **Sort Algorithm** Insertion Sort es correcto.

# Bubble Sort

Representando una implementación con BubbleSort



```python
import random
from termcolor import colored
import copy

def bubble_sort(a, verbose=False):
    n = len(a)
    T = 0 #contador de comparaciones

    if verbose == True: print("input array:",a)

    for i in range(1,n+1):
        if verbose == True: print("\nPasada",i)
        # Flag que indica si el arreglo ya se encuentra ordenado
        already_sorted = True

        # Se recorre el arreglo
        for j in range(n - i):
            T +=1
            if a[j] > a[j + 1]:
                # Si el elemento es mayor al siguiente se intercambia
                a[j], a[j + 1] = a[j + 1], a[j]
                if verbose == True: 
                  print(str(a[:j])[1:-1],",",colored(f"{str(a[j:j+2])[1:-1]}","red"),",",str(a[j+2:])[1:-1])

                already_sorted = False

        if already_sorted == False and verbose == True: 
           print("\nAl finalizar pasada:")
           print(str(a[:n-i])[1:-1],",",colored(f"{str(a[n-i:])[1:-1]}","blue"))

        if already_sorted:
            if verbose == True: print("is sorted!")
            break

    if verbose == True: print("\noutput array:",a)

    return a, T

# Ejemplo
A = [5,8,7,1,3]
print("Entrada: ",A)
A, counter = bubble_sort(A)
print("Salida: ",A)
print("# comparaciones: ", counter)
```

# 5.3. Comparación bubbleSort vs insertionSort
